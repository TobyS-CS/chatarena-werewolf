```python
from chatarena.environments.werewolf.werewolf import Werewolf

# env = Werewolf(player_names=[p.name for p in [player1, player2, player3, player4, player5, player6, player7]])
env = Werewolf(
    player_names=[p.name for p in [player1, player2, player3]],
    role_distribution = (1, 2, 0, 0, 0, 0)
)
```

```python
from chatarena.arena import Arena

# arena = Arena(players=[player1, player2, player3, player4, player5, player6, player7],
arena = Arena(players=[player1, player2, player3],
              environment=env, global_prompt=environment_description)
# Run the game for 10 steps
arena.run(num_steps=10)

# Alternatively, you can run your own main loop
for _ in range(10):
    arena.step()
    # Your code goes here ...
```
